```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def target(x):
    return ((.2*x)**2)*((np.sin(x))**2 +.1)+(np.exp(.05*(x-3)))
```

```python
x=np.arange(-4,4,.01)

figure, axis = plt.subplots()
axis.plot(x,target(x))
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
# pick 5 random points to try out
x=np.random.uniform(-4,4,5)


figure, axis = plt.subplots()

axis.plot(x,target(x),'o')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
# pick a single random starting point to try out
x0=np.random.uniform(-4,4,1)

figure, axis = plt.subplots()

axis.plot(x0,target(x0),'+r')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
# check out two points close on each side of this starting point
epsilon = .2
slope=np.empty(2)
slope[0] = x0+epsilon
slope[1] = x0-epsilon

figure, axis = plt.subplots()

axis.plot(x0,target(x0),'+r')
axis.plot(slope,target(slope),'k')

axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
# the left point has a lower target function value than the starting point 
# - move that direction, look around again
x0 -=epsilon

# the left point has a lower target function value than the starting point 
# - move that direction, look around again
#x0 +=epsilon

# repeat this cell until the slope is close to flat

slope=np.empty(2)
slope[0] = x0+epsilon
slope[1] = x0-epsilon

figure, axis = plt.subplots()
axis.plot(x0,target(x0),'+r')
axis.plot(slope,target(slope),'k')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```

```python
x=np.arange(-4,4,.01)

# plot the true function and the final point.
figure, axis = plt.subplots()
axis.plot(x,target(x))
axis.plot(x0,target(x0),'+r')
axis.plot(slope,target(slope),'k')
axis.set_xlabel("x")
axis.set_ylabel("target(x)")
axis.set_title('target function to minimize');
axis.set_ylim([0.7,1.5])
axis.set_xlim([-4.2,4.2])
```
