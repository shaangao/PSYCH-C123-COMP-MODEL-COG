## Q3.1

Write a function that computes $f(n)=\sum_{i=1}^n 1/(i+1)$. Copy your answer into Gradescope's 3.1 field.

```python
#YOUR CODE HERE
def f(n):
    items = np.array([1 / (i + 1) for i in range(1, n + 1)])
    return np.sum(items)
```

## Q3.2
Plot function $f(n)$ from $n=1$ to $100$.


```python
#YOUR CODE HERE
n = np.arange(1, 101, 1)
y = [f(i) for i in n]

figure, axis = plt.subplots()

#YOUR CODE HERE
axis.plot(n, y)
axis.set_xlabel("$n$")
axis.set_ylabel("$f(n)$")

figure.savefig('PS1_Q3_2.png') 
```

## Q4 
Plot a histogram of $sin(x)$ where $x$ is chosen from a standard normal distribution (mean = 0, sd = 1). Draw 10,000 samples to make the histogram.

```python
#YOUR CODE HERE
samples = np.random.normal(0, 1, 10000)
data = np.sin(samples)


figure, axis = plt.subplots()

#YOUR CODE HERE
axis.hist(data, bins = 100)
axis.set_xlabel("$sin(x)$")
axis.set_ylabel("freq")         

figure.savefig('PS1_Q4.png') 
```
